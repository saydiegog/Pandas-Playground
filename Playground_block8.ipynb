```python
'''
8. Merge & Join en pandas
Conceptos clave

pandas permite combinar DataFrames igual que SQL: INNER, LEFT, RIGHT, OUTER.

merge() se basa en columnas clave; join() se basa en Ã­ndices.

Ideal para integrar datos de diferentes tablas o fuentes.

Tipos de combinaciones

Inner Join

Devuelve solo las filas con coincidencia entre ambos DataFrames.

Ãštil para obtener intersecciones limpias.

Left Join

Mantiene todas las filas del DataFrame izquierdo y agrega coincidencias del derecho.

Cuando quieres preservar el dataset principal aunque no existan match.

Right Join

Igual que left pero tomando como referencia el DataFrame derecho.

Pocas veces necesario; mejor reorganizar tus tablas.

Outer Join

Combina todo, coincida o no.

Celdas sin match quedan como NaN.

ParÃ¡metros importantes en merge()

on: columna(s) clave iguales en ambos DataFrames.

left_on / right_on: cuando las columnas clave tienen nombres distintos.

how: tipo de join (inner, left, right, outer).

suffixes: para evitar colisiones de nombre entre columnas idÃ©nticas.

validate: Ãºtil para verificar integridad relacional ("one_to_one", "one_to_many", etc.).

Uso de join()

Se usa para unir DataFrames por Ã­ndice, no por columnas.

Ãštil cuando ya tienes Ã­ndices significativos o multÃ­ndices.

Sintaxis mÃ¡s simple: df1.join(df2, how="left").

Diferencias rÃ¡pidas

merge() â†’ Une por columnas (mÃ¡s flexible).

join() â†’ Une por Ã­ndices (mÃ¡s directo).


| `merge()` | Cuando unes por columnas | Flexible y recomendado                     |
| `join()`  | Cuando unes por Ã­ndices  | Menos flexible, da errores si no coinciden |
'''
```

```python
import pandas as pd

orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4],
    'customer_id': ['C001', 'C002', 'C003', 'C001'],
    'product_id': ['P01', 'P02', 'P03', 'P02'],
    'quantity': [2, 1, 4, 2]
})

customers = pd.DataFrame({
    'customer_id': ['C001', 'C002', 'C004'],
    'customer_name': ['Ana', 'Luis', 'Marta'],
    'country': ['MÃ©xico', 'PerÃº', 'Chile']
})

products = pd.DataFrame({
    'product_id': ['P01', 'P02', 'P03'],
    'product_name': ['Laptop', 'Mouse', 'Tablet'],
    'price': [1200, 25, 300]
})
```

```python
orders.head()
```

```python
'''
Ejercicio C1 â€” Merge bÃ¡sico (inner)

Une orders + products para agregar el nombre y precio del producto.

ðŸ‘‰ Objetivo: merge con on='product_id'.

Ejercicio C2 â€” Detectar clientes sin pedidos

Haz un left join entre customers y orders y filtra los clientes que no aparecen en orders.

ðŸ‘‰ Objetivo: identificar valores NaN.

Ejercicio C3 â€” AÃ±adir informaciÃ³n del cliente

Une orders + customers para obtener el paÃ­s del cliente.

Pista:

orders.merge(customers, on='customer_id', how='left')

Ejercicio C4 â€” Triple merge

Combina orders + customers + products en un solo DataFrame completo.

ðŸ‘‰ Objetivo: practicar merges encadenados.

Ejercicio C5 â€” Ver quÃ© productos nunca se han vendido

products LEFT JOIN orders, encontrar product_id con NaN.

Ejercicio C6 â€” Merge con mÃºltiples claves

Crea este mini DF para practicar:

extra = pd.DataFrame({
    'customer_id': ['C001','C001','C002'],
    'product_id': ['P01','P02','P02'],
    'rating': [5, 4, 3]
})


Ahora une orders + extra usando
on=['customer_id','product_id'].

Ejercicio C7 â€” Outer merge (todos los datos)

Une customers + orders usando how='outer'.

ðŸ‘‰ Analiza quÃ© filas vienen de cada tabla.

Ejercicio C8 â€” Indicator

Haz un merge con:

how='outer', indicator=True


ðŸ‘‰ Analiza quÃ© filas existen en ambos o solo en uno.'''
```

```python
'''
Ejercicio C1 â€” Merge bÃ¡sico (inner)

Une orders + products para agregar el nombre y precio del producto.

ðŸ‘‰ Objetivo: merge con on='product_id'.'''

orders_products = orders.merge(products, on='product_id')
orders_products.head()
```

```python
print(customers.columns)
print('##########')
print(products.columns)
print('##########')
print(orders.columns)
```

```python
'''
Ejercicio C2 â€” Detectar clientes sin pedidos

Haz un left join entre customers y orders y filtra los clientes que no aparecen en orders.

ðŸ‘‰ Objetivo: identificar valores NaN.'''

customers_no_orders = customers.merge(orders, on='customer_id', how='left')
customers_no_orders = customers_no_orders[customers_no_orders['order_id'].isna()]
print(customers_no_orders)
```

```python
'''
Ejercicio C3 â€” AÃ±adir informaciÃ³n del cliente

Une orders + customers para obtener el paÃ­s del cliente.'''

client_country = orders.merge(customers, on='customer_id', how='left')
print(client_country)
```

```python
'''
Ejercicio C4 â€” Triple merge

Combina orders + customers + products en un solo DataFrame completo.

ðŸ‘‰ Objetivo: practicar merges encadenados.'''

Customer_merged = (
    orders
    .merge(customers, on='customer_id', how='left')
    .merge(products, on='product_id', how='left')
)

print(Customer_merged)
```

```python
'''
Ejercicio C5 â€” Ver quÃ© productos nunca se han vendido

products LEFT JOIN orders, encontrar product_id con NaN.'''

products_never_bought = products.merge(orders, on='product_id', how='left')
products_never_bought = products_never_bought[products_never_bought['order_id'].isna()]
print(products_never_bought)
```

```python
'''
Ejercicio C6 â€” Merge con mÃºltiples claves

Crea este mini DF para practicar:

extra = pd.DataFrame({
    'customer_id': ['C001','C001','C002'],
    'product_id': ['P01','P02','P02'],
    'rating': [5, 4, 3]
})


Ahora une orders + extra usando
on=['customer_id','product_id'].'''

extra = pd.DataFrame({
    'customer_id': ['C001','C001','C002'],
    'product_id': ['P01','P02','P02'],
    'rating': [5, 4, 3]
})

multi_key = orders.merge(extra, on=['customer_id', 'product_id'], how='left')
print(multi_key)
```

```python
'''
Ejercicio C7 â€” Outer merge (todos los datos)

Une customers + orders usando how='outer'.

ðŸ‘‰ Analiza quÃ© filas vienen de cada tabla.'''

outer_customer_order = customers.merge(orders, on='customer_id', how='outer')
print(outer_customer_order)
```

la fila 3 viene de orders
la fila 4 viene de customers

```python
'''
Ejercicio C8 â€” Indicator

Haz un merge con:

how='outer', indicator=True


ðŸ‘‰ Analiza quÃ© filas existen en ambos o solo en uno'''

outer_customer_order2 = customers.merge(orders, on='customer_id', how='outer', indicator=True)
print(outer_customer_order2)

```

Filas 0,1,2 â†’ aparecen en ambas tablas (both)<br>
Fila 3 â†’ aparece solo en orders (right_only)<br>
Fila 4 â†’ aparece solo en customers (left_only)<br>
