```python
'''
Agrupaciones con groupby()

groupby() sirve para dividir el DataFrame en grupos según una o varias columnas, y luego aplicar una operación (sumar, contar, promediar, etc.).
'''
```

```python
'''
Ejercicio 4 — Agrupaciones y Estadísticas
4.1
Agrupa por Segment y muestra las ventas totales (sum de Sales) por segmento.

4.2
Agrupa por Sub-Category y muestra el promedio de Profit por cada subcategoría.

4.3
Filtra todas las filas donde Profit sea negativo,
luego agrupa por City,
suma el Profit por cada ciudad
y muestra las 5 ciudades con la pérdida total más alta.
(Pista: para pérdida "más alta" necesitas TOP 5 de valores más pequeños)

4.4
Agrupa por Product Name, suma las ventas de cada producto
y muestra los 3 productos con mayores ventas.

4.5
Agrupa por Category y muestra el promedio de Discount por categoría.'''

#4.1
print('4.1')
print(df.groupby('Segment')['Sales'].sum())

#4.2
print('4.2')
print(df.groupby('Sub-Category')['Profit'].mean())

#4.3
print('4.3')
df_negative = df[df['Profit'] < 0]
print(df_negative.groupby('City')['Profit'].sum().nsmallest(5))

#4.4
print('4.4')
print(df.groupby('Product Name')['Sales'].sum().nlargest(3))

#4.5
print('4.5')
print(df.groupby('Category')['Discount'].mean())
```

```python
'''
Concepto 1: Crear Columnas Nuevas
En pandas puedes crear columnas nuevas usando operaciones con otras columnas:
➤ Ejemplo 1 — Calcular margen (%)
df["Margin"] = df["Profit"] / df["Sales"]
Esto crea una columna llamada Margin en todo el DataFrame.

Ejemplo 2 — Crear una columna condicional (if…)
Se usa np.where:
import numpy as np
df["High_Discount"] = np.where(df["Discount"] > 0.3, "YES", "NO")

Concepto 2: apply()

apply() te permite aplicar una función personalizada a cada fila o a cada columna.

➤ Ejemplo — Clasificar ventas
def classify(sales):
    if sales > 500:
        return "High"
    elif sales > 100:
        return "Medium"
    else:
        return "Low"

df["Sales_Level"] = df["Sales"].apply(classify)
Útil para reglas más complejas que np.where.

🧩 Concepto 3: groupby con múltiples columnas
Ya dominas el groupby normal.
Ahora vamos con multi-groupby:

➤ Ejemplo
Ventas totales por Segmento y Categoría:
df.groupby(["Segment", "Category"])["Sales"].sum()

El resultado tiene un MultiIndex, dos niveles de agrupación:

Segment     Category
Consumer    Furniture          120000
            Technology         180000
Corporate   Furniture           80000
...

Este tipo de agrupación es clave en análisis más complejos.

🧩 Concepto 4: reset_index()

Cuando haces groupby, pandas te devuelve un índice especial.
Si quieres convertirlo en columnas normales:

➤ Ejemplo
grouped = df.groupby("Category")["Sales"].sum().reset_index()


Output deja de tener MultiIndex y queda como un DataFrame normal.

🧩 Concepto 5: sort_values()

Ya lo has usado con unidimensiones.
Ahora aplicado a DataFrames completos:

df.sort_values("Sales", ascending=False)

🧩 Concepto 6: asignar varias columnas a la vez
df[["col1", "col2"]] = df["Sales"].apply(lambda x: pd.Series([x*2, x*3]))


No lo usarás mucho, pero te da una idea del poder.
'''
```

```python
'''
Ejercicios – BLOQUE 5 (Transformaciones y Grouping Avanzado)


✅ 5.1 — Crear columna “Profit_Margin (%)”

Crea una columna nueva llamada:
Profit_Margin = Profit / Sales
Y muestra las primeras 5 filas.

Concepto base: operaciones columna-columna.

✅ 5.2 — Clasificar productos según su margen
Crear columna Margin_Level:
“High” → Profit_Margin > 0.20
“Medium” → > 0.05 y <= 0.20
“Low” → <= 0.05
Puedes usar:
apply()
o np.where anidados

✅ 5.3 — Ventas totales por Segmento y Categoría

Agrupar por dos columnas:
Segment × Category
Y sumar Sales.
Luego ordenarlo de mayor a menor.

✅ 5.4 — Profit promedio por Año
Crear columna Year a partir de Order Date:
(pista: .dt.year)
Agrupar por Year
Obtener el promedio de Profit

✅ 5.5 — Por subcategoría obtener:
total sales
total profit
número de órdenes (count)
Ordenado por total profit (descendente).
Aquí debes usar:
groupby("Sub-Category").agg(...)

✅ 5.6 — Top 3 combinaciones Segment × Category más rentables

Agrupar por:
["Segment", "Category"]
Sumar Profit.
Ordenar.
Tomar el top 3.'''

print('5.1')
df['Profit_Margin'] = df['Profit'] / df['Sales']
print(df.head(5))

print('5.2')
def level(pm):
    if pm < 0.2:
        return "Low"
    elif pm < 0.5:
        return "Medium"
    else:
        return "High"

df['Margin_Level'] = df['Profit_Margin'].apply(level)
print(df)

print('5.3')
print(df.groupby(['Segment', 'Category'])['Sales'].sum().sort_values(ascending=False))

print('5.4')
df['Year'] = df['Order Date'].dt.year
print(df.groupby('Year')['Profit'].mean())

print('5.5')
print(df.groupby('Sub-Category').agg({'Sales': 'sum', 'Profit':'sum', 'Order ID':'count'}).sort_values(by='Profit', ascending=False))

print('5.6')
print(df.groupby(["Segment", "Category"])['Profit'].sum().sort_values(ascending=False).nlargest(3))

```

```python
'''
6.0 – Conceptos necesarios (con ejemplos)
📌 1. Múltiples condiciones (&, |, ~)
⚠️ Muy importante

En pandas:
& = AND
| = OR
~ = NOT

Y cada condición debe ir entre paréntesis.

✔ Ejemplo
# Filtrar filas donde:
# - La ciudad sea "Seattle"
# - Y las ventas sean mayores a 500
df[(df["City"] == "Seattle") & (df["Sales"] > 500)]

✔ OR
df[(df["City"] == "Seattle") | (df["City"] == "Chicago")]

✔ NOT
df[~(df["Segment"] == "Consumer")]

📌 2. Crear columnas nuevas con condiciones
✔ Ejemplo simple:
df["HighSales"] = df["Sales"] > 1000


Resultado: columna con valores True / False.

✔ Ejemplo con condiciones múltiples
df["ProfitStatus"] = df["Profit"].apply(lambda x: "Ganancia" if x > 0 else "Pérdida")

📌 3. Función apply en columnas

apply permite aplicar tu propia función a cada fila o columna.

✔ Ejemplo columna
df["DiscountLevel"] = df["Discount"].apply(lambda x: "Alto" if x > 0.4 else "Normal")

✔ Ejemplo con función completa
def evaluar_fila(row):
    if row["Profit"] < 0:
        return "Malo"
    elif row["Profit"] < 50:
        return "Ok"
    else:
        return "Excelente"

df["Score"] = df.apply(evaluar_fila, axis=1)


axis=1 → la función se ejecuta por fila.'''
```

```python
'''
🟦 BLOQUE 6 – EJERCICIOS

6.1 — Filtra todas las filas donde:
El Segment sea Consumer
El Discount sea mayor a 0.3
Las Sales sean menores a 100

Muestra solo las primeras 10 filas.
6.2 — Filtra solo los pedidos donde:
La categoría NO sea Technology
Usa ~ (NOT).

6.3 — Filtra todas las filas donde:
La ciudad sea New York City o Los Angeles
Y el Profit sea negativo
Devuelve solo las columnas:
City, Sales, Profit

6.4 — Crea una nueva columna llamada ProfitLabel con las reglas:
Profit > 100 → "Excelente"
Profit entre 0 y 100 → "Bueno"
Profit < 0 → "Pérdida"
Usa apply.

6.5 — Crea una columna llamada Margin que sea:
Margin = Profit / Sales
Luego muestra:
El valor mínimo
El valor máximo
El promedio

6.6 — Filtra todas las filas donde:
Discount > 0.5
Y el margen (Margin) sea negativo
Muestra solo 5 filas.

6.7 — Obtén el promedio de Sales por Segment, pero solo considerando filas donde Profit < 0.
Hint: primero filtras, luego agrupas.

6.8 — Crea una función que reciba una fila y regrese:
"Alto" si Sales > 500 y Profit > 50
"Medio" si Sales > 200
"Bajo" en otro caso
Asignarla a una columna Score2.'''

#6.1
print('6.1')
print(df[(df['Segment'] == 'Consumer') & (df['Discount'] > 0.3) & (df['Sales'] < 100)])

#6.2
print('6.2')
print(df[(~ (df['Category'] == 'Technology'))].head(10))

#6.3
print('6.3')
df_6_3 = df[((df['City'] == 'Los Angeles') | (df['City'] == 'New York City')) & (df['Profit'] < 0)]
print(df_6_3[['City', 'Sales', 'Profit']])

#6.4
print('6.4')

def profit_lbl(profit):
    if profit > 100:
        return "Excelente"
    elif profit > 0 and profit < 100:
        return "Bueno"
    elif profit < 0:
        return "Pérdida"
    else:
        return "Neutro"
    
df['Profit_Label'] = df['Profit'].apply(profit_lbl)
print(df.head(5))

#6.5
print('6.5')
df['Margin'] = df['Profit'] / df['Sales']
print(f"Min:{df['Margin'].nsmallest(1)}")
print(f"Max:{df['Margin'].nlargest(1)}")
print(f"AVG:{df['Margin'].mean()}")

#6.6
print('6.6')
print(df[(df['Discount'] > 0.5) & (df['Margin'] < 0)].head(5))

#6.7
print('6.7')
df_6_7 = df[df['Profit'] < 0]
print(df_6_7.groupby('Segment')['Sales'].mean())

#6.8
print('6.8')
def Score2(row):
    if row['Sales'] > 500 and row['Profit'] > 50:
        return "Alto"
    elif row['Sales'] > 200:
        return "Medio"
    else:
        return "Bajo"

df["Score2"] = df.apply(Score2, axis=1)
print(df.head(10))
```
